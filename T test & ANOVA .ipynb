```python
## T test
l=[23,32,3,5,7,76,6,6,7,88,7,6,6,5,54,4,34,3,3,3,]
assumed_mean=50
from scipy.stats import ttest_1samp
stats,p_value = ttest_1samp(l,assumed_mean)
print(p_value)
if p_value>0.05:
    print("the mean of sample is same as assumed mean")
else:
    print("the mean of sample is not same as assumed mean")
```

```python
l1=[1,2,3,4,5]
l2=[10,20,30,40,50]
from scipy.stats import ttest_ind
p_value = ttest_ind(l1,l2)[1]
print(p_value)
```

```python
l1=[1,2,3,4,5]
l2=[10,20,30,40,50]
from scipy.stats import ttest_ind
p_value = ttest_ind(l1,l2)[1]
print(p_value)
```

```python
if p_value>0.05:
 print("both are same")
else:
  print("not same")
```

```python
n1=[27,29,31,33,35]
n2=[45,50,55,60,65]
from scipy.stats import ttest_rel
stats,p_value = ttest_rel(n1,n2)
print(p_value)
if p_value>0.05:
 print(" marks are same after lecture ")
else:
  print(" marks are not same after lecture ")
```

```python
## anova
l1=[1,2,3,4,5]
l2=[10,20,30,40,50]
l3=[100,200,300,400,500]
from scipy.stats import f_oneway
stats,p_value = f_oneway(l1,l2,l3)
print(p_value)
```

```python
l1 = [12, 14, 16, 18, 20]
l2 = [20, 25, 30, 35, 40]
l3 = [10, 20, 30, 40, 50]
from scipy.stats import stats
stats, p_value = stats.f_oneway(l1,l2,l3)
print(p_value)
if p_value > 0.05:
    print("The means of the groups are  different.")
else:
    print("The means of the groups are not different.")
```
