```python
# Calculate Factorial with For Loop
n = int(input('88: '))
fact = 1
for i in range(1, n + 1):
    fact *= i
print('The factorial of', n, 'is', fact)
```

```python
## calculate factorial with while loop
n=int(input('85:'))
fact = 1
i = 1
while i <=n:
  fact*=i
  i+=1
print(f" the factorial of {n} is {fact}")
```

```python
## check prime number
n=int(input('45:'))
if n >=1:
   print("not prime")

else:
    for i in range (2,int(n**0.5)+1):
        if n % i== 0:
            print("not prime")
            break
    else:
        print("prime")
```

```python
## reverse a number
num=  int(input("59:"))
rev = 0
while  num>0:
   rev = (rev*10)+(num%10)
   num //=10
print(f"reverse number is {rev}")
```

```python
## check if a number is palindrome
num=int(input("19:"))
temp=num
rev=0
while num>0:
    rev=(rev*10)+(num%10)
    num//=10
if temp==rev:
    print("palindrome")
else:
    print("not palindrome")
```

```python
## fibonacci series using for loops
num=int(input("66:"))
a,b=0,1
print(a)
print(b)
for i in range (2,num):
    c=a+b
    print(c)
    a,b=b,c

```

```python
## fibonacci series using while loops
num=int(input("55:"))
a,b=0,1
print(a)
print(b)
i=3
while i <=num:
     c= a+b
     print(c)
     a,b=b,c
     i+=1
```

```python
## Check if a number is Armstrong
n=int(input("11:"))
x=n
sum=0
while n>0:
    digit=n%10
    sum+=digit**3
    n//=10
if x==sum:
   print("armstrong")
else:
   print("not armstrong")
```

```python
## print left triangle
n= int(input("9:"))
for i in range(1,n+1):
    for j in range(1,i+1):
      print("*",end="")
      print()
```

```python
## print righr triangle
n=int(input("22:"))
for i in range(1,n+1):
    for j in range(1<n+1-i):
      print(" ",end="")
      for k in range(1,i+1):
        print("*",end="")
        print()


```
