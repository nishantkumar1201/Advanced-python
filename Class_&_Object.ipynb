```python
### what will be the salary of an individual with 5 years of experience
```

```python
n = len(df)
print(n)
sum_xy= (df['YearsExperience']*df['Salary']).sum()
print('sum_xy',sum_xy)
sumxy= (df['YearsExperience']).sum()*(df['Salary']).sum()
print('sumxy',sumxy)
```

```python
numerator = n*sum_xy - sumxy
numerator
```

```python
sumx= (df['YearsExperience']).sum()
print('sumx', sumx)


sumx2= (df['YearsExperience']).sum()**2
print('sumx2',sumx2)


sum_x2= (df['YearsExperience']**2).sum()
print('sum_x2',sum_x2)

sumy = (df['Salary']).sum()
print('sumy',sumy)

slope= ((n*sum_xy)-sumxy)/((n*sumx2)-sum_x2)
print("slope:",slope)

intercept= (sumy- slope*sumx)/n
print("intercept:",intercept)

print('y=',slope,'x+',intercept)

sal = slope*5 + intercept
print('salary of an individual with 5 years of experience:',sal)
```
