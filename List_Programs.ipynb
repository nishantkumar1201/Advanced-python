```python
# List creation
```

```python
course=["Python","SQL","Stats","power BI"]
print(course)
```

```python
course.append("data science essential")
print(course)
```

```python
# Insert Items
# To insert a list item at a specified index, use the insert() method
```

```python
course.insert(3,'photography')
print(course)
```

```python
company=['bmw','kawasaki ','ducati']
print(company)
```

```python
# Extend List
# To append elements from another list to the current list, use the extend() method.
```

```python
course.extend(company)
```

```python
print(course)
```

```python
# Remove Specified Item
# The remove() method removes the specified item.
```

```python
course.remove('photography')
print(course)
```
