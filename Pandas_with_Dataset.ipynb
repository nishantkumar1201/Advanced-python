```python
import pandas as pd
import numpy as np

df1=pd.read_csv('/content/Health_data.csv')
```

```python
df1
```

```python
df1.head()
```

```python
#rename(), which lets  change column names.
df1.rename(columns={'Duration': 'Timing'})
```

```python
#rename() lets you rename index values by specifying a index

df1.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
df2=pd.read_csv('/content/sales.csv')
```

```python
df2.head()
```

```python
pd.concat([df1, df2])
```

```python
#join() lets you combine different DataFrame objects which have an index in common.

df1.join(df2)
```
