## Python Class Inheritance Example

```
class Vehicle:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def display_info(self):
        print(f"Brand: {self.brand}, Color: {self.color}")

class Car(Vehicle):
    def __init__(self, brand, color, seats):
        super().__init__(brand, color)
        self.seats = seats

    def display_info(self):
        print(f"Brand: {self.brand}, Color: {self.color}, Seats: {self.seats}")

# Testing
car = Car("Toyota", "Red", 5)
car.display_info()
```

## PyTorch Tensor Operations

```
import torch

# 1. Create the tensors
zeros_tensor = torch.zeros((3, 3))
ones_tensor = torch.ones((2, 4))
random_tensor = torch.rand((2, 2))

print("Zeros Tensor:\n", zeros_tensor)
print("Ones Tensor:\n", ones_tensor)
print("Random Tensor:\n", random_tensor)
```

```
# 2. Create tensor from 1 to 6, reshape, convert to float, then to NumPy
values = torch.arange(1, 7)
reshaped = values.view(2, 3)
reshaped_float = reshaped.float()
numpy_array = reshaped_float.numpy()

print("NumPy array:\n", numpy_array)
```
