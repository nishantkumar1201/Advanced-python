```python
print(set1)
```

**Joining Sets**

```python
#Union Method
set1= {"x","y","z"}
set2={6,7,8}
set3=set1.union(set2)
print(set3)
```

```python
# | Operator
set1= {"x","y","z"}
set2={6,7,8}
set3= set1 | set2
print(set3)
```

```python
#Join mutiple sets
set1= {"x","y","z"}
set2={6,7,8}
set3={'Motu','Patlu'}
set4={'King','kobra'}
largeset=set1.union(set2,set3,set4)
print(largeset)
```

```python
set1= {"x","y","z"}
set2={6,7,8}
set3={'Motu','Patlu'}
set4={'King','kobra'}
largeset=set1 |set2 |set3 |set4
print(largeset)
```

```python
#update Method
set1= {"x","y","z"}
set2={6,7,8}
set1.update(set2)
print(set1)
```

```python
#intersection method
set1= {"x","y","z"}
set2={6,7,8,"x"}
set3= set1.intersection(set2)
print(set3)
```

```python
# & operator
set1= {"x","y","z"}
set2={6,7,8,"x"}
set3= set1 & set2
print(set3)
```

```python
# difference method
set1= {"x","y","z"}
set2={6,7,8,"x"}
set3= set1.difference(set2)
print(set3)
```

```python
set1= {"x","y","z"}
set2={6,7,8,"x"}
set3= set2.difference(set1)
print(set3)
```
