```python
def decor_result(result_function):
    def distinction(marks):
      for m in marks:
        if m>=75:
              print("distinction")
              break
        else:
          result_function(marks)
    return distinction

@decor_result
def result(marks):
  for m in marks:
    if m>33:
      pass
    else:
      print("fail")
      break
  else:
    print("pass")

result([50,60,70,30,90,80])
```
