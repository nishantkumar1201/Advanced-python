```python
from scipy.stats import shapiro
```

```python
sub_df1=df['sepal_length']
```

```python
stat,p=shapiro(sub_df1)
print(" test stat is:",stat)
print("p value is:",p)
if p>0.05:
    print("data is normally distributed")
else:
    print("data is not normally distributed")
```
