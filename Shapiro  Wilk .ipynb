 **SHAPIRO WILK TEST**

 HO: data is normaly distributed
 HA: data is not normaly distributed



```python
from scipy.stats import shapiro
data = [1.2,1.4,1.6,1.8,2.0,2.2,2.4,2.6,2.8,3.0]
stats,p= shapiro(data)
print("test stats is:", stats)
print("p-value is:",p)

if p>0.05:
  print("data is normaly distributed")
else:
  print("data is not normaly distributed")
```

```python
data = [1.2,1.4,1.6,1.8,2.0,2.2,2.4,2.6,2.8,3.0,500,600,560]
stats,p= shapiro(data)
print("test stats is:", stats)
print("p-value is:",p)

if p>0.05:
  print("data is normaly distributed")
else:
  print("data is not normaly distributed")
```

```python
import numpy as np
from scipy.stats import pearsonr

# example data
x= [1,2,3,4,5]
y= [2,4,6,8,10]

# pearson correlation
r,p= pearsonr(x,y)


print("Pearsonr correlation Test:")
print("correlation  coefficient :",r)
print("p-value :",p)

# interpretation
if p>0.05:
  print(" the data is lilely not correlated (fail to reject HO).")
else:
  print("the data is correlated (reject HO).")
```

```python
from scipy.stats import spearmanr

# example data
x= [1,2,3,4,5]
y= [5,6,7,8,7]

# spearman correlation
rho, p = spearmanr(x,y)
print("spearman correlation test:")
print("correlation coefficient (p):", rho)
print(f"p-value :",p)

# interpretation
if p>0.05:
  print(" the data is lilely not correlated (fail to reject HO).")
else:
  print("the data is correlated (reject HO).")

```
