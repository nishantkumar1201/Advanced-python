# TensorFlow Tensor and Math Operations

```
import tensorflow as tf
```

## Create Tensors

```
# Create a constant 1-D tensor `a` with values [10, 20, 30, 40]
a = tf.constant([10, 20, 30, 40])
print("Tensor a:", a)
```

```
# Create another constant 1-D tensor `b` with values [2, 4, 6, 8]
b = tf.constant([2, 4, 6, 8])
print("Tensor b:", b)
```

## Mathematical Operations

```
# Element-wise addition of a and b
add_result = tf.add(a, b)
print("Element-wise Addition:", add_result)
```

```
# Element-wise multiplication of a and b
mul_result = tf.multiply(a, b)
print("Element-wise Multiplication:", mul_result)
```

```
# Dot product of a and b
dot_result = tf.tensordot(a, b, axes=1)
print("Dot Product:", dot_result)
```
